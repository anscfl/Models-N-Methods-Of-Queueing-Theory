```python
mu_1 = 10
mu_2 = 8
lambd = 15
phi = 1

alpha = lambd / (mu_1 + mu_2)
beta = mu_2 / mu_1

temp_1 = alpha / (1 + 2 * alpha)
temp_2 = 1 / beta
temp_3 = 1 + (1 + beta ** 2) * alpha - (1 - beta ** 2) * phi

p_00 = (1 - alpha) / (1 + temp_1 * temp_2 * temp_3)
print('Вероятность того, что оба продовца свободны:', p_00)

# вероятность того, что первый продавец занят обслуживанием покупателей, а второй свободен (и наоборот)
p_10 = temp_1 * (1 + beta) * (alpha + phi) * p_00
print('Вероятность того, что первый продавец занят, а второй свободен:', p_10)
p_01 = temp_1 * ((1 + beta) / beta) * (alpha + 1 - phi) * p_00
print('Вероятность того, что первый свободен, а второй продавец занят:', p_01)
# доля рабочего времени, когда оба продовца будут свободны (заняты)
p_busy = 1 - p_00 - (p_10 + p_01)
print('Доля р.в., когда оба продовца заняты =', p_busy)
p_free = p_00
print('Доля р.в., когда оба продовца свободны =', p_free)

# вероятность того, что в магазине больше 3 покупателей
p_2 = (alpha ** 2 / (1 + 2 * alpha)) * ((1 + beta) / beta) * (1 + (1 + beta) * alpha - (1 - beta) * phi) * p_00
p_3 = (alpha ** 3 / (1 + 2 * alpha)) * ((1 + beta) / beta) * (1 + (1 + beta) * alpha - (1 - beta) * phi) * p_00
p_m = 1 - (p_00 + p_01 + p_10 + p_2 + p_3)
print('Вероятность того, что в магазине больше 3 покупателей:', p_m)

# среднее число покупателей в магазине
temp_1 = alpha * (1 + beta) / (1 - alpha)
temp_2 = 1 + (1 + beta) * alpha - (1 - beta) * phi
temp_3 = beta * (1 + 2 * alpha) + alpha * (1 + (1 + beta ** 2) * alpha - (1 - beta ** 2) * phi)

n = temp_1 * (temp_2 / temp_3)
print('Среднее число покупателей в магазине:', n)

# среднее время ожидания покупателя в очереди
h = (p_01 + p_10) + 2 * p_busy
b = n - h
w = b / lambd
print('Среднее время ожидания покупателя в очереди:', w)

# среднее время пребывания покупателя в магазине
u = n / lambd
print('Среднее время пребывания покупателя в магазине:', u)
```

```python
from math import exp, factorial
import numpy as np

N = 2
lam = 1.5
nu = 0.5
mu = 1 / nu

alpha = lam / mu
print('alpha:', alpha)


def p1n(_n):
    return (alpha * sum([alpha ** (-i) for i in range(N - _n + 1)])) / ((N + 1) * exp(alpha) + sum([(N - i + 1) * alpha ** (-i) for i in range(1, N+1)]))


def pkn(_k):
    return ((alpha ** _k) / factorial(_k)) / ((N + 1) * exp(alpha) + sum([(N - i + 1) * alpha ** (-i) for i in range(1, N+1)]))


# а) вероятность того, что у раздат. окна инструм. кладовой никого из рабочих нет, и дежурный кладовщик свободен от выдачи
p_1_0 = p1n(0)
p_0_0 = p_1_0 / alpha
print('Вероятность того, что у раздат. окна инструм. кладовой никого из рабочих нет, и дежурный кладовщик свободен от выдачи:\n', p_0_0)

# б) вероятности того, что на раздаче инструмента будет занято 1,2,3,...,7 кладовщиков
p_1_1 = p1n(1)
p_1_2 = p1n(2)
p = [
    [p_1_0],
    [p_1_1],
    [p_1_2]
]

for i in range(N+1):
    for k in range(2, 8):
        p[i].append(pkn(k))
        
p = np.array(p)
p_k = [p.transpose()[k].sum() for k in range(7)]

print('\nВероятности того, что на раздаче инструмента будет занято 1,2,3,...,7 кладовщиков:\n')
for k in range(7):
    print('p(', k + 1, ', n) = ', round(p_k[k], 5), ', где n = 0,1,2.', sep='')
    
# в) вероятности того, что очередь не будет превышать N человек при условии,
# что на раздаче инструмента будет занято 1,2,3,...,7 кладовщиков
print('\nВероятности того, что очередь не будет превышать N человек при условии, что на раздаче инструмента будет занято 1,2,3,...,7 кладовщиков:\n')

p_N = [p_0_0] * 7
for n in range(N + 1):
    for k in range(7):
        p_N[k] += p[n][k]

[print("Вероятности того, что очередь не будет превышать 2 человек при условии, что на раздаче инструмента будет занято", k + 1, 'кладовщиков:', round(p_N[k], 4)) for k in range(len(p_N))]

# г) среднее число занятых кладовщиков
h = sum([p1n(n) for n in range(N + 1)]) + sum([k * sum([pkn(k) for _ in range(N + 1)]) for k in range(2, 150)])
print('М.о. числа занятых кладовщиков:', h)

# д) среднее число рабочих, ожидающих в очереди
b = sum([n * p1n(n) for n in range(N + 1)]) + sum([sum([n * pkn(k) for n in range(N + 1)]) for k in range(2, 150)])
print('М.о. числа рабочих, ожидающих в очереди:', b)

# е) среднее время пребывания рабочих в очереди в ожидании выдачи инструмента
w = b / lam
print('М.о. времени пребывания рабочих в очереди:', w)

# среднее число кладовщиков с условием
a = 80 / 100
s = p_0_0
k = 0
while s < a:
    s += p_k[k]
    k += 1

print(f'В среднем необходимо {k} кладовщиков, чтобы обеспечить обслуживание рабочих в заданных условиях с гарантийной веоятностью не ниже 0.8')
```
